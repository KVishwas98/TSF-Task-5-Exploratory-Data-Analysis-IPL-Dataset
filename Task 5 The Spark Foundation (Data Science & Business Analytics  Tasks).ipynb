```python
# let's visualize match data by 'season' using countplot

plt.figure(figsize=(12,6))
sns.countplot(x="season", data= matches)
plt.xticks(rotation = "vertical")
plt.show()
```

```python
# let's visualize  match data by 'winner' using barplot

plt.figure(figsize=(12,6))
data = matches.winner.value_counts()
sns.barplot(x=data.index, y= data, orient= "v")
plt.xticks(rotation = "vertical")
plt.show()
```

### 'Mumbai Indians' has won the most tournament followed by 'Chennai Super Kings' and 'Kolkatta Knight Riders'.

```python
# let's visualize  match data by 'venue' using countplot

plt.figure(figsize=(12,6))
sns.countplot(x="venue", data = matches)
plt.xticks(rotation = "vertical")
plt.show()
```

### 'Rajiv Gandhi International Stadium, Uppal' and Maharashtra Cricket Assosciation Stadium are mostly used for IPL matches. 

```python
# let's visualize  match data by 'win_by_runs' using swarmplot

plt.figure(figsize=(20,10))
sns.swarmplot(y="win_by_runs", x = "season" ,data = matches)
plt.xticks(rotation = 90)
plt.show()
```

## With this above plot we can interpret most matches have been won in the range of 20-60 runs.

```python
# let's visualize  match data by 'toss_decision' using countplot

plt.figure(figsize=(12,6))
sns.countplot(x="season", hue = "toss_decision",  data = matches)
plt.xticks(rotation = "vertical")
plt.show()
```

## Team's mostly have chosed to bowl when they had won the toss, and in the recent years from 2016 to 2019 choosing bowl has become a default option.

```python
# let's find out the team's which all have won the match who have chosed to bowl first.

num_of_wins = (matches.win_by_wickets>0).sum()
num_of_loss = (matches.win_by_wickets==0).sum()
labels = ["Wins", "Loss"]
total = float(num_of_wins + num_of_loss)
sizes = [(num_of_wins/total)*100, (num_of_loss/total)*100]
colors = ["orange", "pink"]
plt.pie(sizes, labels = labels, colors = colors,
       autopct = "%1.1f%%", shadow= True, startangle=90)
plt.title("Win Percentage Batting Second")
plt.show()
```

## As we can see team's which have chosed to bowl first won the match and percentage of winning the match is 53.7%.

```python
# Will Explore the performace of the batsman and find out who were the best performace in the IPL.

best_p = deliveries.groupby("batsman")["batsman_runs"].agg("sum").reset_index().sort_values(by = "batsman_runs", ascending = False).reset_index(drop=True)
best_p = best_p.iloc[:10,:]

labels = np.array(best_p["batsman"])
ind = np.arange(len(labels))
width = 0.7
fig, ax = plt.subplots(figsize=(15,8))
rects = ax.bar(ind, np.array(best_p["batsman_runs"]), width = width, color = "red")
ax.set_xticks(ind+((width)/2.))
ax.set_xticklabels(labels, rotation = "vertical")
ax.set_ylabel("Count")
ax.set_title("Batsman - Top Run Scorers in IPL")
ax.set_xlabel("Batsman's Name")


plt.show()

```

### dammn!! 'Virat Kohli' has scored the highest in the history of IPL and then 'Suresh Raina' followed by 'RG Shamma'. till 2009.

# Getting exciting here!

```python
## Let's see the number of 6's

best_s = deliveries.groupby("batsman")["batsman_runs"].agg(lambda x: (x==6).sum()).reset_index().sort_values(by="batsman_runs", ascending= False).reset_index(drop=True)
best_s = best_s.iloc[:10,:]

labels = np.array(best_s["batsman"])
ind = np.arange(len(labels))
width = 0.8
fig, ax = plt.subplots(figsize=(15,8))
rects = ax.bar(ind, np.array(best_s["batsman_runs"]), width = width, color = "pink")
ax.set_xticks(ind+((width)/2.))
ax.set_xticklabels(labels, rotation = 90)
ax.set_ylabel("Count")
ax.set_title("Batsman - Most No. of SIXES in IPL")
ax.set_xlabel("Batsman's Name")


plt.show()
```

## Bammn!! 'Chris Gayle' has shot more sixes followed by 'AB de Villiers' and 'MS Dhoni'.

```python
## Let's see the number of 4's

best_f = deliveries.groupby("batsman")["batsman_runs"].agg(lambda x: (x==4).sum()).reset_index().sort_values(by="batsman_runs", ascending= False).reset_index(drop=True)
best_f = best_f.iloc[:10,:]

labels = np.array(best_f["batsman"])
ind = np.arange(len(labels))
width = 0.6
fig, ax = plt.subplots(figsize=(15,8))
rects = ax.bar(ind, np.array(best_f["batsman_runs"]), width = width, color = "blue")
ax.set_xticks(ind+((width)/2.))
ax.set_xticklabels(labels, rotation = 90)
ax.set_ylabel("Count")
ax.set_title("Batsman - Most No. of FOURS in IPL")
ax.set_xlabel("Batsman's Name")


plt.show()
```

## 'Shikhar Dhawan' has hit most four followed by 'Suresh Raina' and then 'Gautam Gambhir'.

```python
# let's visualize  match data by 'batting team' using countplot

plt.figure(figsize=(12,6))
sns.countplot(x="batting_team" , data = deliveries)
plt.xticks(rotation = "vertical")
plt.show()
```

### 'Mumbai Indians', Kolkatta Knight Riders and Kings XI Punjab won most matches where they have chosed to bat at first.

### Here we see the batting team deliveries where 'Kolkatta Knight Riders' led the chart.

```python
# Let's find out the performance of the best bowler in the IPL. 

plt.figure(figsize=(12,6))
sns.countplot(x="dismissal_kind" , data = deliveries)
plt.xticks(rotation = "vertical")
plt.show()
```

### Caught wicket is the only reason which knocked out batsman the most followed by bowled and then run out.

```python
# ## Let's see the bowler with Extra Runs

best_b = deliveries.groupby("bowler")["extra_runs"].agg(lambda x: (x==0).sum()).reset_index().sort_values(by="extra_runs", ascending= False).reset_index(drop=True)
best_b = best_b.iloc[:10,:]

labels = np.array(best_b["bowler"])
ind = np.arange(len(labels))
width = 0.6
fig, ax = plt.subplots(figsize=(15,8))
rects = ax.bar(ind, np.array(best_b["extra_runs"]), width = width, color = "blue")
ax.set_xticks(ind+((width)/2.))
ax.set_xticklabels(labels, rotation = 90)
ax.set_ylabel("Count")
ax.set_title("Bowlera - with Extra Runs")
ax.set_xlabel("Bowler's Name")


plt.show()
```

### Hurrey!! 'Harbhajan Singh' and then 'PP Chawla' were on the top hit list followed by 'A Mishra' and 'R Ashwin'.

```python
# ## Let's see the bowler with Total Runs

best_b = deliveries.groupby("bowler")["total_runs"].agg(lambda x: (x==0).sum()).reset_index().sort_values(by="total_runs", ascending= False).reset_index(drop=True)
best_b = best_b.iloc[:10,:]

labels = np.array(best_b["bowler"])
ind = np.arange(len(labels))
width = 0.6
fig, ax = plt.subplots(figsize=(15,8))
rects = ax.bar(ind, np.array(best_b["total_runs"]), width = width, color = "green")
ax.set_xticks(ind+((width)/2.))
ax.set_xticklabels(labels, rotation = 90)
ax.set_ylabel("Count")
ax.set_title("Bowlera - with Total Runs")
ax.set_xlabel("Bowler's Name")


plt.show()
```

### Hurrey!! 'Harbhajan Singh' and then 'SL Malinga' were on the top hit list followed by 'B Kumar' and 'A Mishra'.
